```python
#importing library, and loading the dataset from a local directory into a dataframe
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
df=pd.read_csv("spotify-2023.csv", encoding="latin-1")
```

```python
#df["streams"]=pd.to_numeric(df["streams"])
#Shows error caused by the 574 index.
df.loc[574]
```

```python
#Instead of dropping it, changed its value so it isn't causing errors
df.loc[574,"streams"]=0
```

```python
#Merging columns / Changing datatypes
#Date should be one column, and key+mode as well
df["streams"]=pd.to_numeric(df["streams"])
df[df["track_name"].duplicated()]
```

```python
df["key"]=df["key"]+" "+df["mode"]
```

```python
df["release_date"] = pd.to_datetime(df["released_year"].astype(str) + "-" +
                                    df["released_month"].astype(str) + "-" +
                                    df["released_day"].astype(str))
```

```python
list(df)
```

```python
#Dropping the excess columns after merging
df.drop(columns=["released_month","released_day","mode"], inplace=True)
```

```python
df.duplicated().sum()
```

```python
df = df.sort_values(by="streams", ascending=False)
```

```python
df
```

```python
df.isnull().sum()
#I decide to fill up the Empty null values, so i can avoid dropping important datas with them
```

```python
df["key"].fillna("Unknown", inplace=True)
```

```python
df["in_shazam_charts"].fillna(0, inplace=True)
```

```python
df.info()
```

```python
#df["in_deezer_playlists"] = pd.to_numeric(df["in_deezer_playlists"])
#Shows an error caused by commas used as a decimal separator
```

```python
df["in_deezer_playlists"] = df["in_deezer_playlists"].str.replace(',', '')
```

```python
df["in_deezer_playlists"] = pd.to_numeric(df["in_deezer_playlists"])
#df["in_shazam_charts"] = pd.to_numeric(df["in_shazam_charts"])
#Same comma error as previously
```

```python
df["in_shazam_charts"] = df["in_shazam_charts"].str.replace(',', '')
```

```python
df["in_shazam_charts"] = pd.to_numeric(df["in_shazam_charts"])
```

```python
df.info()
```

```python
#Investigating why the column is in a float format
df["in_shazam_charts"]
```

```python
#Removing NaN values, and coverting the column
df["in_shazam_charts"].fillna(0,inplace=True)
```

```python
df["in_shazam_charts"]=df["in_shazam_charts"].astype(int)
```

Ready to visualize

```python
result = df.groupby("released_year").agg({"streams": "max", "track_name": "first", "artist(s)_name":"first"})
```

```python
result_df=pd.DataFrame(result)
```

```python
result_df.sort_values(by="streams", ascending=False)
```
